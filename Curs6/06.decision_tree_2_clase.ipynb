```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("doua_clase.csv")
x = df[["coord_x", "coord_y"]]
y = df["category"]
```

```python
plt.scatter(df["coord_x"], df["coord_y"]);
```

```python
d_colors = {
    "a":"red", "b":"black", "c":"blue", "d":"green"
}
```

```python
plt.scatter(df["coord_x"], df["coord_y"], c=df["category"].replace(d_colors));
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model
```

```python
model.fit(x, y)
```

```python
model.predict([[200, 100]])
```

```python
from sklearn.tree import export_graphviz
import graphviz
```

```python
exported_data = export_graphviz(model, feature_names=["coord_x", "coord_y"], class_names=["a", "b", "c", "d"], filled=True, rounded=True)
```

```python
graph = graphviz.Source(exported_data)
graph.render("4 Categories", format="png")
graph.view()
```
